```python
from scipy.stats import spearmanr

feature_cols = ["liquidity_z", "mom_12_1_z", "mom_6_1_z", "rev_1m_z", "vol_12m_z", "beta_12m_z"]

rows = []

for col in feature_cols:
    tmp = ml_df[["date", col, "fwd_ret_1m"]].dropna().copy()

    monthly_ic = (
        tmp.groupby("date")
        .apply(lambda x: spearmanr(x[col], x["fwd_ret_1m"]).statistic)
        .reset_index(name="rank_ic")
    )

    rows.append({
        "feature": col,
        "mean_rank_ic": monthly_ic["rank_ic"].mean(),
        "std_rank_ic": monthly_ic["rank_ic"].std(),
        "ir_rank_ic": monthly_ic["rank_ic"].mean() / monthly_ic["rank_ic"].std(),
        "positive_ic_pct": (monthly_ic["rank_ic"] > 0).mean(),
        "n_months": len(monthly_ic),
    })

feature_ic_summary = pd.DataFrame(rows).sort_values("mean_rank_ic", ascending=False)
feature_ic_summary
```
