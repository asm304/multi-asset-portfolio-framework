```python
from src.backtest.engine import build_monthly_returns

monthly_returns = build_monthly_returns()

full_eval = alpha.merge(
    monthly_returns,
    on=["date", "ticker"],
    how="left"
)

full_eval.groupby("selected")["fwd_ret_1m"].describe()
```

```python
selected_mean = full_eval.loc[full_eval["selected"], "fwd_ret_1m"].mean()
not_selected_mean = full_eval.loc[~full_eval["selected"], "fwd_ret_1m"].mean()

selected_mean - not_selected_mean
```

```python
full_eval = full_eval.dropna(subset=["alpha", "fwd_ret_1m"]).copy()

full_eval["alpha_decile"] = pd.qcut(
    full_eval["alpha"],
    q=10,
    labels=False,
    duplicates="drop"
)

decile_returns = full_eval.groupby("alpha_decile")["fwd_ret_1m"].mean()
decile_returns
```

```python
full_eval = full_eval.dropna(subset=["alpha", "fwd_ret_1m"]).copy()

full_eval["alpha_decile_cs"] = (
    full_eval.groupby("date")["alpha"]
    .transform(lambda x: pd.qcut(x, q=10, labels=False, duplicates="drop"))
)

decile_returns_cs = full_eval.groupby("alpha_decile_cs")["fwd_ret_1m"].mean()
decile_returns_cs
```

```python
from scipy.stats import spearmanr

feature_cols = ["liquidity_z", "mom_12_1_z", "mom_6_1_z", "rev_1m_z", "vol_12m_z", "beta_12m_z"]

rows = []

for col in feature_cols:
    tmp = ml_df[["date", col, "fwd_ret_1m"]].dropna().copy()

    monthly_ic = (
        tmp.groupby("date")
        .apply(lambda x: spearmanr(x[col], x["fwd_ret_1m"]).statistic)
        .reset_index(name="rank_ic")
    )

    rows.append({
        "feature": col,
        "mean_rank_ic": monthly_ic["rank_ic"].mean(),
        "std_rank_ic": monthly_ic["rank_ic"].std(),
        "ir_rank_ic": monthly_ic["rank_ic"].mean() / monthly_ic["rank_ic"].std(),
        "positive_ic_pct": (monthly_ic["rank_ic"] > 0).mean(),
        "n_months": len(monthly_ic),
    })

feature_ic_summary = pd.DataFrame(rows).sort_values("mean_rank_ic", ascending=False)
feature_ic_summary
```
